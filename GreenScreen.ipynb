```python
import matplotlib.pyplot as plt
import matplotlib.image as mpimg
# import matplotlib.pyplot as plt
import numpy as np
import cv2

%matplotlib inline
```

```python
image = mpimg.imread('alienGreenScreen .jpeg')
print('Image dimensions:', image.shape)
```

```python
plt.imshow(image)
```

```python
lower_green = np.array([0,100,0])
upper_green = np.array([100,255,100]) # Green should be large to make it white
```

```python
# Define the masked area
mask = cv2.inRange(image, lower_green, upper_green)

# Vizualize the mask
plt.imshow(mask, cmap='gray')
```

```python
masked_image = np.copy(image) # copy the image first

masked_image[mask != 0] = [0,0,0]

plt.imshow(masked_image)
```

```python
# Load in a background image, and convert it to RGB
background_image = mpimg.imread('4-space.jpeg')

background_image = cv2.cvtColor(background_image, cv2.COLOR_BGR2RGB)

crop_background = background_image[0:480, 0:852]

# Mask the cropped background so that the pizza area is blocked
crop_background[mask == 0] = [0, 0, 0]

# Display the background
plt.imshow(crop_background)
```

```python
# Add two images together to create a complete image
complete_image = masked_image + crop_background

plt.imshow(complete_image)
```
